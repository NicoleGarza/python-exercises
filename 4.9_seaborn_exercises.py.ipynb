# Use the iris database to answer the following questions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset('iris')
```

# What does the distribution of petal lengths look like?

```python
sns.distplot(iris.petal_length)
```

# Is there a correlation between petal length and petal width?

```python
sns.relplot(x='petal_length', y='petal_width', data=iris)
```

# Would it be reasonable to predict species based on sepal width and sepal length?

```python
sns.relplot(x='petal_length', y='petal_width',hue='species', data=iris)
```

# Which features would be best used to predict species?

```python
sns.pairplot(iris)
```

# Create a column named is_catholic that holds a boolean value of whether or not the province is Catholic. (Choose a cutoff point for what constitutes catholic)

```python
is_catholic = swiss.Catholic > 50
swiss['is_catholic']=is_catholic # creates new column for above
swiss
```

# Does whether or not a province is Catholic influence fertility?

```python
swiss[['is_catholic','Fertility']].groupby('is_catholic').mean()
```

```python
sns.heatmap(swiss.corr(),annot=True)
```

# Using the chipotle dataset from the previous exercise, create a barplot that shows the 4 most popular items and the revenue produced by each.

```python
from sqlalchemy import create_engine
from env import user, host, pw
url = 'mysql+pymysql://{}:{}@{}/chipotle'.format(user,pw, host)
dbc=create_engine(url)
df = pd.read_sql('SELECT * FROM orders',dbc)
df.head()
```

```python
price_no_dollarsign = df.item_price.replace('[\$,]', '', regex=True).astype(float)
df['price_no_dollarsign']=price_no_dollarsign
most_popular = df[['price_no_dollarsign', 'item_name']].groupby('item_name',as_index=False).sum().sort_values(by='price_no_dollarsign', ascending=False).head(4)

most_popular

sns.barplot(data=most_popular, y='price_no_dollarsign', x= 'item_name')
```
