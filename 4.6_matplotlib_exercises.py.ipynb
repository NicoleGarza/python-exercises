# Use matplotlib to plot the following equation: y=x^2−x+2

```python
import matplotlib.pyplot as plt
```

```python
x = list(range(-49, 50))
y = [n ** 2 - n + 2 for n in x]
plt.scatter(x, y)
plt.title('An Exponential Distribution, $y = x^2 - x + 2$')
plt.xlabel('$x$')
plt.ylabel('$x^2 - x + 2$')
plt.show()
```

# Add an annotation for the point 0,0 the origin

```python
x = list(range(-49, 50))
y = [n ** 2 - n + 2 for n in x]
plt.scatter(x, y)
plt.title('An Exponential Distribution, $y = x^2 - x + 2$')
plt.xlabel('$x$')
plt.ylabel('$x^2 - x + 2$')
plt.annotate('(0 , 0)', xy=(0, 0), xytext=(0, 500),
             arrowprops={'facecolor': 'blue'})
plt.show()
```

# Create and label 4 separate plots for the following equations (choose a range for x that makes sense): y=√x,  y=x3, y=tan(x), y=2x

```python
import math

plt.figure(figsize=(16, 6))

x = list(range(1, 100))
y = [math.sqrt(n) for n in x]
z = [n ** 3 for n in x]
n = [math.tan(n) for n in x]
r = [2 ** n for n in x]

plt.subplot(141)
plt.plot(x, y)
plt.title('y=√x')

plt.subplot(142)
plt.plot(x, z)
plt.title('y=x^3')

plt.subplot(143)
plt.plot(x, n)
plt.title('y=tan(x)')

plt.subplot(144)
plt.plot(x, r)
plt.title('y=2^x')

plt.suptitle('Subplots')

plt.show()
```

# Comine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend.

```python
import math

plt.figure(figsize=(16, 6))

x = list(range(1, 100))
y = [math.sqrt(n) for n in x]
z = [n ** 3 for n in x]
n = [math.tan(n) for n in x]
r = [2 ** n for n in x]

plt.plot(x, y, c='yellow', label='$√x$')

plt.plot(x, z, c='blue', label='$x^3$')

plt.plot(x, n, c='purple', label='$tan(x)$')

plt.plot(x, r, c='red', label='$2^x$')

plt.title('Combos')
plt.legend(loc='upper right')

plt.ylim(-200, 200)
plt.xlim(0, 100)

plt.show()
```
